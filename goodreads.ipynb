```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Lectura y vusialización del dataset

df = pd.read_csv('books.csv', error_bad_lines = False)
df
```

```python
# Información del dataset
df. info()
df.describe(include ='all')
```

```python
# Libros con mayor número de reviews

most_reviewed = df.sort_values('text_reviews_count', ascending = False).head(10).set_index('title')
plt.figure(figsize=(15,10))
sns.barplot(most_reviewed['text_reviews_count'], most_reviewed.index, alpha=.9,palette='cubehelix')
plt.show()
```

```python
# Libros con mejor puntuación

best_rated = df.sort_values('average_rating', ascending = False).head(10).set_index('title')
plt.figure(figsize = (15, 10))
sns.barplot(best_rated['average_rating'], best_rated.index, alpha = .9, palette = 'cubehelix')
plt.show()
```

```python
#Libros con mejor rating

most_rated = df.sort_values('ratings_count', ascending = False).head(10).set_index('title')
plt.figure(figsize = (15, 10))
sns.barplot(most_rated['ratings_count'], most_rated.index, alpha = .9, palette = 'cubehelix')
plt.show()
```
